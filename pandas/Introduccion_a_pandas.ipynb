# Introducción a Pandas
Pandas es una librería que proporciona estructuras de datos y herramientas de análisis de datos de alto rendimiento y fáciles de usar.
* La estructura de datos principal es el DataFrame, que puede considerarse como una tabla 2D en memoria (como una hoja de cálculo, con nombres ded columna y etiquetas de fila).
* Muchas funciones disponibles en Excel están disponibles mediante programación, como crear tablas dinamicas, calcular columnas basadas en otras columnas, trazar graficos, etc.
* Proporciona un alto rendimiento para manipular (Unir, dividir, modificar...) grandes conjuntos de datos.

### Import

```python
import pandas as pd
```

## Estructuras de datos en Pandas
La librería pandas, de manera genérica, contiene las siguientes estructuras de datos:
* Series: Array de una dimension
* DataFram: Se corresponde con una tabla de 2 dimensiones
* Panel: Similar a un diccionario de DataFrames.

## Creación del objeto Series

```python
# Creación de un objeto Series
s = pd.Series([2, 4, 6, 8, 10]) 
print(s)
```

```python
# Creación de un objeto Series inicializádolo
# con un diccionario de Python
altura = {"Santiago": 187, "Pedro": 178, "Julia": 170, "Ana": 165}
s = pd.Series(altura)
print(s)
```

```python
# Creación de un objeto Series inicializándolo con algunos
# de los elementos de un diccionario de Python
altura = {"Santiago": 187, "Pedro": 178, "Julia": 170, "Ana": 165}
# reducir diccionario
s = pd.Series(altura, index=["Pedro", "Julia"]) 
print(s)
```

```python
# Creación de un objeto Series inicializandolo
# con un escalar
s = pd.Series(34, ["test1", "test2", "test3"])
print(s)
```

## Acceso a los elementos de un objeto series
Cada elemento en un objeto Series tiene un identificador único que se denomina <strong>index label</strong>


```python
# Creación de un objeto Series
s = pd.Series([2, 4, 6, 8], index=["num1", "num2", "num3", "num4"])
print(s)
```

```python
# Accediendo al tercer elemento del objeto
s["num3"]
```

```python
# Tambien se puede acceder al elemento por posición
s[3]
```

```python
# loc es la forma estándar de acceder a un elemento
# de un objeto Series por atributo o etiqueta
s.loc["num1"]
```

```python
# iloc es la forma estándar de acceder a un elemento 
# de un objeto Series por posición
s.iloc[0]
```

```python
# Slice
s.iloc[2:4]
```

# Dataframes
### Creación de un objeto DataFrame


```python
# Creación de un DataFrame inicializándolo con un diccionario de objetos Series
personas = {
    "peso": pd.Series([84,90,56,64],["Santiago", "Samuel", "Sebastián", "Sara"]),
    "altura": pd.Series({"Santiago": 187, "Samuel": 178, "Sebastián": 170, "Sara": 165}),
    "hijos": pd.Series([2,3], ["Sebastián", "Samuel"])
}
df = pd.DataFrame(personas)
df
```

```python
# Creación de un DataFrame inicializándolo con algunos elementos de un diccionario
# de objetos Series
personas = {
    "peso": pd.Series([84,90,56,64],["Santiago", "Samuel", "Sebastián", "Sara"]),
    "altura": pd.Series({"Santiago": 187, "Samuel": 178, "Sebastián": 170, "Sara": 165}),
    "hijos": pd.Series([2,3], ["Sebastián", "Samuel"])
}

df = pd.DataFrame( personas, 
                   columns=["altura", "peso"],
                   index = ["Sebastián", "Samuel"])
df
```

```python
# Creación de un DataFrame inicializándolo con una listas de Python
# Importante: Deben especificarse las columnas e indices por separado
valores = [
    [185,4,75],
    [170,0,65],
    [190,1,89]
]

df = pd.DataFrame(valores, 
                  columns=["altura", "hijos","peso"], 
                  index=["Sebastian", "Samuel", "Santiago"])
df
```

```python
# Manera mas sencilla de crear un Dataframe
# Con un diccionario
personas = {
    "altura": {"Santiago": 187, "Sebastian": 178, "Sara": 156},
    "peso": {"Santiago": 47, "Sebastian": 68, "Sara": 56}
}
df = pd.DataFrame(personas)
df
```

## **Modificación de un DataFrame**

```python
# Creación de un DataFrame inicializándolo con un diccionario de objetios Series
personas = {
    "peso": pd.Series([84, 90, 56, 64], ["Santiago","Pedro", "Ana", "Julia"]),
    "altura": pd.Series({"Santiago": 187, "Pedro": 178, "Julia": 170, "Ana": 165}),
    "hijos": pd.Series([2, 3], ["Pedro", "Julia"])
}

df = pd.DataFrame(personas)
df
```

```python
# Añadir una nueva columna al DataFrame
df["cumpleaños"] = [1990,1987, 1980, 1994]
df
```

```python
# Añadir una nueva columna calculada al DataFrame
df["años"] = 2020 - df["cumpleaños"]
df
```

```python
# Añadir una nueva columna creando un DataFrame nuevo
df_mod = df.assign(mascotas = [1,3,0,0])
print(df_mod)
print(df)
```

```python
# Eliminar una columna existente del DataFrame
del df["peso"]
df
```

```python
# Eliminar una columna existente devolviendo una copia del DataFrame resultante
df_mod = df.drop(["Ana"], axis=0) # Eje X
df_mod = df_mod.drop(["años"], axis=1) # Eje Y
df_mod
```

## **Evaluación de expresiones en un DataFrame**

```python
# Creación de un DataFrame inicializándolo con un diccionario de objetios Series
personas = {
    "peso": pd.Series([84, 90, 56, 64], ["Santiago","Pedro", "Ana", "Julia"]),
    "altura": pd.Series({"Santiago": 187, "Pedro": 178, "Julia": 170, "Ana": 165}),
    "hijos": pd.Series([2, 3], ["Pedro", "Julia"])
}

df = pd.DataFrame(personas)
df
```

```python
# Evaluar una función sobre una columna del DataFrame
df.eval("altura / 2")
```

```python
# Asignar el valor resultante como una nueva columna
df.eval("media_altura = altura / 2", inplace=True)
df
```

```python
# Evaluar una función utilizando una variable local
max_altura = 180
df.eval("altura > @max_altura")
#df[df.eval("altura > @max_altura")]
```

```python
# Aplicar una función externa a una columna del DataFrame
def func(x):
  return x + 2
df["peso"] = df["peso"].apply(func)
df
```
